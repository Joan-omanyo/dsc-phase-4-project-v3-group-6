### Clean Text (Essential for Sentiment Analysis)
Now we clean the tweets: remove URLs, mentions, hashtags, punctuation, etc.

Cleaning the tweet text, is essential because raw tweets often contain noise such as URLs, hashtags, mentions, punctuation, numbers, and inconsistent capitalization, which can distort analysis. Cleaning standardizes the text, making it easier for algorithms and visualizations to correctly identify patterns, such as frequently used words, sentiment-related phrases, or common topics. Without this step, irrelevant characters could bias word counts, wordclouds, or machine learning models, leading to inaccurate insights

```python
def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www.\S+", "", text)      # URLs
    text = re.sub(r"@\w+", "", text)                # @mentions
    text = re.sub(r"#\w+", "", text)                # hashtags
    text = re.sub(r"[^a-zA-Z\s]", "", text)         # punctuation/numbers
    text = re.sub(r"\s+", " ", text).strip()        # spaces
    return text

df["clean_text"] = df["tweet_text"].apply(clean_tweet)
```

### Tokenization

Split each tweet into words (needed for stop word removal, stemming, TF-IDF, or Bag-of-Words):

```python
from nltk.tokenize import word_tokenize
import nltk
df['tokens'] = df['clean_text'].apply(word_tokenize)
```

### Stop Word Removal

Remove common words like “the”, “and”, but keep negations like “not” for sentiment:

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) - set(['not', 'no'])
df['tokens'] = df['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```

### Stemming

Reduce words to their root/base form:

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
# Stemming
df['tokens'] = df['tokens'].apply(lambda x: [stemmer.stem(word) for word in x])
```

After tokenization, stop word removal, and stemming, we can join the tokens back into a single string. This is useful for visualizations like wordclouds or frequency analysis while keeping the text clean and consistent.

```python
df['clean_text_final'] = df['tokens'].apply(lambda x: ' '.join(x))
```

Automated Feedback Categorisation

A custom categorisation function grouped each tweet into one of the following product-relevant categories:

Hardware (e.g., battery, screen, camera)

Software (e.g., app crashes, interface issues)

Price

Service (e.g., store experience, staff interactions)

General/Other

The categorisation was applied to the dataset, and a frequency analysis was performed specifically for negative tweets to identify the most problematic areas.

Business relevance:

Helps determine whether issues are predominantly hardware-related, software-driven, pricing concerns, or service complaints.

Supports prioritisation of engineering resources.

Provides product teams with clear direction on improvement areas.

```python
# 2. Categorization Logic
def categorize_feedback(text):
    text = text.lower()

    if any(x in text for x in ['battery', 'screen', 'button', 'case', 'camera']):
        return 'Hardware'
    elif any(x in text for x in ['app', 'crash', 'interface', 'update', 'software', 'map']):
        return 'Software'
    elif any(x in text for x in ['price', 'cost', 'expensive', 'cheap']):
        return 'Price'
    elif any(x in text for x in ['store', 'service', 'genius', 'staff']):
        return 'Service'
    else:
        return 'General/Other'


def build_feedback_categories(df):
    df['category'] = df['text'].apply(categorize_feedback)

    print("\nFeedback Categories Distribution (Negative Tweets):")
    neg_tweets = df[df['sentiment'] == 'Negative']
    print(neg_tweets['category'].value_counts())

    return df
```
